A **character-level language model** is predicting the next character in a sequence given some concrete sequence of characters before it.

Note that every single word (name) in the above list contains multiple examples.

E.g. the word 'emma' contains the information:

- 'e' is likely to come first in a sequence of a name.
- 'm' is likely to come after 'e'
- 'm' is likely to come after 'em'
- 'a' id likely to come after 'emm'
- After 'emma' the sequence is very likely to end.

A **bi-gram language model** only looks at one character (the immediate previous one) in the given sequence to predict the next character.

```python
for w in words[:1]:
    for ch1, ch2 in zip(w, w[1:]):
        print(ch1, ch2)
```

But recall that we have more information packed here: the head and the tail!

To treat them, we define a "start" token before the head, and an "end" token after the tail:

```python
for w in words[:1]:
    chars = ['<S>'] + list(w) + ['<E>']
    for ch1, ch2 in zip(chars, chars[1:]):
        print(ch1, ch2)
```
